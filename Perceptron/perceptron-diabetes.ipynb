## Aquire the Diabetes data
pandas will be used to fall data manipulation

```python
import pandas as pd

df = pd.read_csv('diabetes.csv')

df.tail()
```

### Extract the first 500 class labels corresponding to Outcome

```python
import matplotlib.pyplot as plt
import numpy as np
y = df.iloc[0:500, 8].values
y = np.where(y == 0, -1, 1)

print(y)
```

### Extract the first 500 class labels skin thickness and BMI

```python
x = df.iloc[0:500, [3, 5]].values

print(x)
```

## Visualize the data on a Scatter Plot

```python
plt.scatter(x[:250, 0], x[:250, 1],color='red', marker='o', label='skin thickness')
plt.scatter(x[250:500, 0], x[250:500, 1], color='blue', marker='x', label='BMI')

plt.xlabel('skin thickness')
plt.ylabel('BMI')
plt.legend(loc='upper left')
plt.show()
```

## Train the Perceptron on the Diabetes data collected

### Graph the misclassification errors agains the number of epochs

```python
from perceptron import Perceptron

pn = Perceptron(0.1, 500)
pn.fit(x, y)

plt.plot(range(1, len(pn.errors) + 1), pn.errors, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of misclassifications')
plt.show()
```

### Vistualize the decision boundaries

Create a function that will help with the creation of shading the coloring the decision boundaries

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
   # setup marker generator and color map
   markers = ('s', 'x', 'o', '^', 'v')
   colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
   cmap = ListedColormap(colors[:len(np.unique(y))])

   # plot the decision surface
   x1_min, x1_max = X[:,  0].min() - 1, X[:, 0].max() + 1
   x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
   xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))

   Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
   Z = Z.reshape(xx1.shape)
   
   plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
   plt.xlim(xx1.min(), xx1.max())
   plt.ylim(xx2.min(), xx2.max())

   # plot class samples
   for idx, cl in enumerate(np.unique(y)):
      plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
      alpha=0.8, c=cmap(idx),
      marker=markers[idx], label=cl)
```

Plot the decision regions on the previous scatterplot

```python
plot_decision_regions(x, y, classifier=pn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```
