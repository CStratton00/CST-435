```python
# import the appropriate libraries
# https://www.tensorflow.org/tutorials/load_data/csv

import pandas as pd
import numpy as np

# Make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

import warnings
warnings.filterwarnings("ignore") #suppress warnings

import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, mutual_info_regression, f_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# load in the dataset
basketball_seasons = pd.read_csv('all_seasons.csv')
basketball_seasons.head()
```

```python
fields = ['age', 'player_height', 'player_weight', 'gp', 'pts', 'reb', 'ast', 'net_rating', 'oreb_pct', 'dreb_pct', 'usg_pct', 'ts_pct', 'ast_pct']
relevant_stats = pd.read_csv('all_seasons.csv', usecols=fields)

X = relevant_stats.drop(columns=['net_rating']).to_numpy()
y = relevant_stats['net_rating'].to_numpy()

selector = SelectKBest(mutual_info_regression, k=5)
best_feat = selector.fit_transform(X, y)
sorted_indicies = np.argsort(selector.scores_)
```

```python
fields = ['gp', 'pts', 'reb', 'net_rating']
relevant_stats = pd.read_csv('all_seasons.csv', usecols=fields)

basketball_labels = relevant_stats.pop('net_rating')
relevant_stats = np.array(relevant_stats).astype('float32')

basketball_model = tf.keras.Sequential([
  layers.Dense(128),
  layers.Dense(1)
])

basketball_model.compile(loss = tf.losses.MeanSquaredError(),
                            optimizer = tf.optimizers.Adam())

basketball_model.fit(relevant_stats, basketball_labels, epochs=15)
```

```python
normalize = preprocessing.Normalization()
normalize.adapt(relevant_stats)

norm_abalone_model = tf.keras.Sequential([
  normalize,
  layers.Dense(64),
  layers.Dense(1)
])

norm_abalone_model.compile(loss = tf.losses.MeanSquaredError(),
                           optimizer = tf.optimizers.Adam())

norm_abalone_model.fit(relevant_stats, basketball_labels, epochs=10)
```
